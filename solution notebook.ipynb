## 1. What is SQL?

SQL stands for **Structured Query Language**. It's a programming language used to fetch and transform data.

Underneath this text is a SQL cell - it contains SQL code, and executes it upon demand. You run it by clicking 'Run' on the right hand side of the cell. Don't worry about the contents of this cell, just run it.
### Press Run on the SQL Cell below

```r
SELECT
	animal_type,
	COUNT(*) AS incident_count
FROM 'animal_rescue_incidents.csv'
WHERE location_type = 'Outdoor'
GROUP BY animal_type
ORDER BY incident_count DESC
LIMIT 5;
```

SQL is used for querying data sets, so the input and output are always data sets. They have a specific structure:
* The columns of the data set are called 'Fields'. `animal_type` and `incident_count` in the above output data set are `fields`.
* The rows of the data set are sometimes called 'Observations'. DataLab puts an index next to each observation, labeling them 1 to 5 in this case.

## 3. Filtering Data
Now that we know what the data looks like, we can use a filter to focus on the parts of the data that we are most interested in.

### Ask the AI to show you the incidents that were about cats

Answer: <br>
Show me the incidents in 'animal_rescue_incidents.csv' where animal_type is 'cat'

```r
SELECT *
FROM 'animal_rescue_incidents.csv'
WHERE animal_type = 'cat';
```

The AI should have used a `WHERE` clause. `Where` is used alonside a true/false statement to filter the data to cases where that statement is true.

![where image](where.png)

> ðŸ” Troubleshooting: Check that the casing of 'cat' (as opposed to 'Cat') matches the way it is presented in data.

### Ask the AI to show you the incidents that weren't about cats

Answer: <br>
Show me the incidents in 'animal_rescue_incidents.csv' where animal_type is not 'cat'

```r
SELECT *
FROM 'animal_rescue_incidents.csv'
WHERE animal_type <> 'cat';
```

Out of 13,072 total rows there are 6,884 `WHERE animal_type = 'cat'` and 5,923 `WHERE animal_type != 'cat'`. Not everything in this data is either a Cat or not a Cat!
### Ask the AI to show the incidents where we don't know the animal_type.

Answer: <br>
Show me the incidents in 'animal_rescue_incidents.csv' where we don't know the animal_type

```r
SELECT *
FROM 'animal_rescue_incidents.csv'
WHERE animal_type IS NULL OR animal_type = '';
```

There are some entries where the animal type is `NULL`. This isn't text. It's a special type of unknown value that doesn't equal anything. That means it won't show up when you filter for Cats, or exclude Cats, unless you mention it specifically.

> ðŸ” Troubleshooting: Make sure that the code is specifically checking `animal_type IS NULL`, rather than looking for equality such as `animal_type = 'null'` or `animal_type = 'unknown'`.

### Ask the AI to show you incidents where it wasn't a cat, _or_ we don't know what it was.

## 4. Aggregation

So far we've only looked at the data. It's time to start calculating.

### Ask the AI to tell you how many rows there are in the data.

Answer: <br>
How many rows are there in 'animal_rescue_incidents.csv'?

```r
SELECT COUNT(*) FROM 'animal_rescue_incidents.csv';
```

`COUNT(*)` simply tells us the number of rows in the data.

### Ask the AI to tell you how many incidents involved a dog.

Answer: <br>
How many rows are there in 'animal_rescue_incidents.csv' where animal_type is 'dog'?

```r
SELECT COUNT(*) AS dog_incident_count
FROM 'animal_rescue_incidents.csv'
WHERE animal_type = 'dog';
```

The AI may have named the count, perhaps `dog_incident_count`. This is called 'Aliasing', and it's when we specify the name (or 'alias) that a field will take in the output data set.
<br>
<br>
It's helpful to know how many incidents involved a dog, but what if we want a full breakdown of incidents by animal type? Would we have to run a separate piece of code for each one? What if we don't know all the types? There's a simpler solution.
### Ask the AI to get you a count of incidents by animal_type.

Answer: <br>
Get me a count of rows in 'animal_rescue_incidents.csv' by animal_type

```r
SELECT animal_type, COUNT(*) AS incident_count
FROM 'animal_rescue_incidents.csv'
GROUP BY animal_type;
```

This is an aggregation. The `GROUP BY animal_type` clause establishes that we want to separate the table into each `animal_type`, and then we run `COUNT(*)` over each segment.

> ðŸ” Troubleshooting: Make sure that the field is correctly referred to as `animal_type` in both the `SELECT` and `GROUP BY` clauses.

We can aggregate by more than one field at once.
### Ask the AI to count combinations of `location_type` and `rescue_type`.

Answer: <br>
Get me a count of rows in 'animal_rescue_incidents.csv' by location_type and rescue_type

```r
SELECT
  location_type,
  rescue_type,
  COUNT(*) AS incident_count
FROM
  'animal_rescue_incidents.csv'
GROUP BY
  location_type,
  rescue_type;
```

That should show us every combination of `location_type` and `rescue_type` in the table, and a count of how many times it ocurrs. When checking the AI's SQL, notice that just like the the `SELECT` clause, the elements of the `GROUP BY` clause should be separated by commas.

> ðŸ” Troubleshooting: Make sure that the AI has used the correct field names, `location_type` and `rescue_type`, in both the `SELECT` and `GROUP BY` clauses.

## 5. Ranking & Sorting

Did you notice that the data was in random order in the last output? Or did the AI sort it for you? SQL code won't sort the output unless it was specified to do so.
### Ask the AI to show you a count of rows by animal_type sorted by the number of incidents.

Answer: <br>
Show me a count of rows in 'animal_rescue_incidents.csv' by animal_type ordered by the number of incidents

```r
SELECT animal_type, COUNT(*) AS incident_count
FROM 'animal_rescue_incidents.csv'
GROUP BY animal_type
ORDER BY incident_count DESC;
```

The `ORDER BY` clause tells SQL to sort, or order, the output. The AI may or may not have added `DESC`. Without it, the data will be sorted in ascending order - ie. 1, 2, 3 and so on. Adding `DESC` will instead sort the data in descending order, so 3, 2, 1.
<br>
<br>Let's go one step further
### Ask the AI to show you the top 5 Animal Types by number of instances

Answer: <br>
Show me a the top 5 animal_type entries in 'animal_rescue_incidents.csv' by number of incidents

```r
SELECT animal_type, COUNT(*) AS incident_count
FROM 'animal_rescue_incidents.csv'
GROUP BY animal_type
ORDER BY incident_count DESC
LIMIT 5;
```

The AI should have used:
* `SELECT` and `FROM` to set up the query
* `COUNT(*)` and `GROUP BY` to aggregate the incidents
* `ORDER BY`, `DESC` and `LIMIT` to fetch only the top entries

> ðŸ” Troubleshooting: As always, check that the AI has used the correct names for the table and fields.

<br>This is a complex SQL query, but it's easy to conceptualise and ask the AI to write for you.
